#### Введение

Рассмотрим критерии, которые можно использовать для сравнения и выявления взаимосвязи качественных характеристик.

Такого рода задачи возникают в случае изучения предпочтения людей или выявления зависимости между лекарством и выживаемостью.

Предположим, что мы наблюдаем $n$ объектов

$O_1, ..., O_n$ 

Объектом может быть человек, некоторый предмет, например, автомобиль, может быть ответ человека.

Будем предполагать, что объект обладает двумя качественными признаками.

$A: A_1, ..., A_r$

$B: B_1, ..., B_s$

```
pd.DataFrame({'B1': ['n11', 'n21', '...', 'nr1', 'n1'], 'B2': ['n12', 'n22', '...', 'nr2', 'n2'], '...': ['...', '...', '...', '...', '...'], 'Bs': ['n1s', 'n2s', '...', 'nrs', 'ns'], '': ['m1', 'm2', '...', 'mr', 'n'] }, index = ['A1', 'A2', '...', 'Ar', ''])
```

$m_i$ кол-во объектов обладающих значением $A_i$ признака $A$, независимо от того, какое значение принимает признак $B$.

$n_i$ кол-во объектов обладающих значением $B_i$ признака $B$, независимо от того, какое значение принимает признак $A$.

$n$ общее число рассматриваемых объектов.


Такого рода представление наблюдений для дальнейшего изучения взаимосвязи двух признаков, называется **таблица сопряжённости**.

#### Независимость признаков

Обозначим через $p_i$ вер-ть того, что признак $A$ у случайно взятого объекта, принял значение $A_i$.

$p_i = P(A_i), i = \overline {1, r}$

$q_i = P(B_j), j = \overline {1, s}$

$A$ и $B$ называются независимыми, если при любых $i$ и $j$:

$p_{ij} = P(A_i \cap B_i) = p_iq_j$

$p_{ij}$ вер-ть того, что случайно взятый объект имеет одновремено значение признака $A$ равное $A_i$ и признака $B$ равное $B_i$.

#### Гипотеза

$H_0: P(A_i \cap B_j) = p_iq_j$ для любых $i, j$

$H_1: \exists(i, j): P(A_i \cap B_j) \neq p_iq_j$ для любых $i, j$

#### Статистика

Воспользуемся для проверки гипотезы о независимости 2 признаков, критерием хи-квадрат. Будем сравнивать наблюдаемые частоты и ожидаемые.

$\chi^2 = \sum \limits_{i=1}^r \sum \limits_{j=1}^s \frac{(n_{ij} - np_iq_j)^2}{np_iq_j} $, где

$n_{ij}$ наблюдаемые частоты (эмпирические)

$np_iq_j$ ожидаемые частоты (теоритические)

Вычислить истинные значения ожидаемых частот мы не можем, т.к. мы не знаем какая ф-я распределения. Но мы можем найти оценки $p_i$ и $q_j$ с помощью метода максимального правдоподобия. Составим ф-ю правдоподобия 

$L= \frac{n!}{\prod\limits_{i = \overline {1,r} \\ j = \overline {1,s}} n_{ij!}} \prod\limits_{i = \overline {1,r} \\ j = \overline {1,s}} (p_i q_j)^{n_{ij}}$

$\sum \limits_{i=1}^r p_i = 1$ и $\sum \limits_{j=1}^s q_j = 1$

Точечные оценки

$\hat{p_i} = \frac{m_i}{n}, i = \overline {1,r}$

$m_i$ кол-во объектов обладающих значением $i$ из $r$, признаки расположенные по строкам.

$\hat{q_j} = \frac{n_j}{n}, j = \overline {1, s}$

$n_i$ кол-во объектов обладающих значением $i$ из $s$, признаки расположенные по столбцам.

#### Статистический критерий

Подставим в статистику, оценки вероятностей и получим статистический критерий

$\chi^2 = \sum \limits_{i=1}^r \sum \limits_{j=1}^s \frac{(n_{ij} - \frac{m_in_j}{n})^2}{\frac{m_in_j}{n}} \sim \chi^2_{(s-1) \cdot (r - 1)}$, где

$n_{ij}$ наблюдаемые частоты (эмпирические)

$\frac{m_in_j}{n}$ оценка веростности

$(s-1) \cdot (r - 1)$ произведение степеней свободы, где

$r$ кол-во признаков по строкам

$s$ кол-во признаков по столбцам

#### Scipy

На выходе первым идёт критерий, вторым p-value, третьим степени свободы, четвертым ожидаемые частоты, по которым проверяем условие $\frac{m_in_j}{n} \geq 5$

```
obs = np.array([[10, 10, 20], [20, 20, 20]])
st.chi2_contingency(obs)
```

#### Критическая область

$V_k = (\chi^2_{1 - \alpha, (s-1) \cdot (r - 1)}, + \infty)$, где

$(s-1) \cdot (r - 1)$ произведение степеней свободы, где

$r$ кол-во признаков по строкам

$s$ кол-во признаков по столбцам

$\chi^2_в \in V_k \Rightarrow H_0$ отвергается.

#### Статистический критерий. Поправка Йейтса

В случае таблицы размерности 2х2, статистика критерия не редко принимает завышенные значения и следовательно, мы отвергаем нулевую гипотезу чаще, чем предполагали, т.е. не с вер-ю равную уровню значимости $\alpha$.

$\chi^2 = \sum \limits_{i=1}^r \sum \limits_{j=1}^s \frac{(|n_{ij} - \frac{m_in_j}{n}| - 0.5)^2}{\frac{m_in_j}{n}}$

#### Критическая область. Поправка Йейтса

$V_k = (\chi^2_{1 - \alpha, 1}, + \infty)$, где

$\chi^2_{1 - \alpha, 1}$ имеет 1 степень свободы

#### Важное условие!

Ожидаемые частоты для каждой клетки таблицы, должны быть больше 5.

$\frac{m_in_j}{n} \geq 5$

В случае, если в какой-то клетке данное условие не выполняется, нам нужно объединить соседние значения признака.

$\chi^2 = \sum \limits_{i=1}^r \sum \limits_{j=1}^s \frac{(n_{ij} - \frac{m_in_j}{n})^2}{\frac{m_in_j}{n}} \sim \chi^2_{(s-1) \cdot (r - 1)}$

#### Что делать, когда не выполнено условие $\frac{m_in_j}{n} \geq 5$. Точный критерий Фишера

Предположим, что есть $n$ объектов

$O_1, ..., O_n$

Каждый из объекта обладает двумя качественными признаками

$A: A_1, A_2$

$B: B_1, B_2$

```
pd.DataFrame({'': ['A1', 'A2', 'сумма'], 'B1': ['n11', 'n21', 'n1'], 'B2': ['n12', 'n22', 'n2'], 'сумма': ['m1', 'm2', 'n']})
```

#### Гипотеза

Ответим на вопрос, зависимы ли признаки $A$ и $B$?

$H_0: A$ и $B$ зависимы

$H_1: A$ и $B$ независимы

Будем считать, что какой-то из $n_{ij} < 5$. В противном случае, мы бы могли воспользоваться критерий хи-квадрат, в данном случае хи-квадрат использовать мы не можем.

#### Точный критерий Фишера

Основан на сравнении вероятностей реализации таблиц с такими же суммарными значениями по строкам и столбцам, т.е. $n_j$ и $m_i$, что и исходная наблюдаемая таблица.

$T_0$ обозначим так исходную таблицу, так же добавим индекс с ноликом в таблицу. И зафиксируем $n_1, n_2, m_1, m_2$.

```
pd.DataFrame({'': ['A1', 'A2', 'сумма'], 'B1': ['n_0_11', 'n_0_21', 'n1'], 'B2': ['n_0_12', 'n_0_22', 'n2'], 'сумма': ['m1', 'm2', 'n']})
```

При фиксированных $n_1, n_2, m_1, m_2$ рассмотрим все возможные заполнения таблицы. Обозначим одну из таких реализаций, через $T_i$, найдём вероятность получить таблицу $T_i$

```
pd.DataFrame({'': ['A1', 'A2', 'сумма'], 'B1': ['n_i_11', 'n_i_21', 'n1'], 'B2': ['n_i_12', 'n_i_22', 'n2'], 'сумма': ['m1', 'm2', 'n']})
```

$P_i = \frac{C^{n^i_{11}, n^i_{12}, n^i_{21}, n^i_{22}}_n}{C^{n_1, n_2}_n \cdot C^{m_1, m_2}_n} = \frac{\cdot n_{1}! \cdot n_{2}! \cdot m_{1}! \cdot m_{2}!}{n^i_{11}! \cdot n^i_{12}! \cdot n^i_{21}! \cdot n^i_{22}! \cdot n! }$

#### Статистика критерия Фишера 

Обозначим через $P_0$ вероятность реализации таблицы $T_0$ (исходной таблицы по нашим наблюдениям).



Статистика двухстороннего критерия

$P = \sum \limits_{T_i: P_i \leq P_0, i \neq 0} P_i + P_0$

это вероятность равная сумме вероятностей, всех таблиц $T_i$ для которых вероятность реализации, не превосходит заданной вероятности $P_0$.

Если $P < \alpha \Rightarrow H_0$ отклоняем. Значит признаки зависимы!

#### Scipy

Подаём в массив значения из изначальной таблицы. pvalue и есть статистика критерия $P = \sum \limits_{T_i: P_i \leq P_0, i \neq 0} P_i + P_0$

```
oddsratio, pvalue = st.fisher_exact([[8, 2], [1, 5]])
oddsratio, pvalue
```

$P = P_0 + P_1 + P_2 + P_3 + P_4 = 1$

$P > \alpha \Rightarrow H_0$ не отвергается. Оценки 1 и 2 эксперта независимы.

Найдём вер-ть $P_0$ реализации данной таблицы $T_0$



$P_i = \frac{\cdot n_{1}! \cdot n_{2}! \cdot m_{1}! \cdot m_{2}!}{n^i_{11}! \cdot n^i_{12}! \cdot n^i_{21}! \cdot n^i_{22}! \cdot n! }$

$P_0 = \frac{4!3!3!4!}{1!2!3!1!7!}$

```
def count_prob(a, b):
  s = 1
  r = 1
  for x in a:
    s *= math.factorial(x)

  for x in b:
    r *= math.factorial(x)

  return s/r
```

```
a = [4, 3, 3, 4]
b = [1, 2, 3, 1, 7]

p0 = count_prob(a, b)
p0
```
