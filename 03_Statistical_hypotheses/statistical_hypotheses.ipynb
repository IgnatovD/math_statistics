#### Статистический критерий

**Правило проверки статистической гипотезы** при некоторой фиксированной альтернативе называется **статистическим критерием**.

$S$ - критерий

$S = \left\{ Z > с \right\} $ - правило, по которому мы отклоняем гипотезу.

$Z$ - статистика критерия, $c$ - пороговое значение

$H_0$ отвергается $\Leftrightarrow T(x) \geq c$

**Статистикой критерия** значимости называется статистика, по значениям которой судят о справедливости выдвинутой гипотезы.

$z_в = Z$

$Z$ - статистика критерия, например, $Z = \overline X$

$z_в$ - реализация статистики критерия, например, $z = 40$

#### Критическая область

**Уровень значимости $\alpha$**  называется столь малая вероятность, что событие с такой вероятностью является практически невозможным. 

**Критической область критерия** называется подобласть области значений статистики критерия, вероятность попадания в которую для этой статистики при условии истинности гипотезы $H_0$ равна уровню значимости. т.е. если суждение истино, то маловероятно, что статистика может оказаться в критическуой области.

$V$ - область всех возможных значений статистики $Z$.

$V_k$ - критическая область критерия, если статистика $z_в \in V_k$ мы отклоняем $H_0$.

$V_k: P(z_в \in V_k|H_0) = \alpha$  при условии, что $H_0$ верна.

$V$ \ $V_k$ - область принятия $H_0$.

$V$ \ $V_k:$ $P(z_в \notin V_k|H_0) = 1 - \alpha$

#### Ошибка I и Ошибка II рода

**Ошибка I рода** - отклонить гипотезу $H_0$, когда она верна.

Вероятность Ошибки I рода обозначим через $\alpha$

$\alpha = P(z_в \in V_k|H_0)$  при условии, что верна $H_0$

**Ошибка II рода** - приняли гипотезу $H_0$, когда верна $H_1$.

Вероятность Ошибки II рода обозначим через $\beta$

$\beta = P(z_в \in V \backslash V_k|H_1)$ при условии, что верна $H_1$

#### Мощность критерия

**Мощность критерия** - вероятность $\gamma = 1 - \beta = P(z_в \in V_k|H_1)$ один минус вероятность Ошибки II рода. Чем больше мощность, тем меньше вероятность допустить Ошибку II рода.

**Мощность критерия** — это вероятность отклонения нулевой гипотезы, когда она ложна.

**Задача построения оптимального критерия**

\begin{cases}
  \alpha \leq \alpha_0 \\
  \beta \rightarrow  \underset{S}{ min}
\end{cases}

Вероятность Ошибки I рода $\alpha$ не превосходит определенного значения $\alpha_0$, а вероятность Ошибки II рода $\beta$ минимизируем по критерию $S$.

#### p-value

**p-value** называется статистика, равная вероятности того, что статистика критерия примет значение такое же как $z(X_{[x]})$ или более экстремальное при условии, что верна нулевая гипотеза.

$H_0$ отвергается $\Leftrightarrow p_{value} \leq \alpha$

В этом случае $p_{value}$ - степень уверенности в отвержении $H_0$ (чем $p_{value}$ меньше, тем увереннее).

Пусть задана некоторая генеральная совокупность $\xi$, функция распределения сл. вел. $\xi$ нам известна $F_{\xi}(x,\theta)$. Для определенности будем считать, что $\theta$ это одномерный параметр.

Требуется проверить гипотезу

$H_0: \theta = \theta_0$

В зависимости от условий задач, мы можем выбрать одну из следующих альтернативных гипотез.

$H^1_1: \theta = \theta_1 \neq \theta_0$

$H^2_1: \theta > \theta_0$

$H^3_1: \theta < \theta_0$

$H^4_1: \theta \neq \theta_0$

Выберем статистику критерия. **В качестве статистики критерия предлагается использовать точечную оценку неизвестного параметра**.

Пусть $Z = \hat{\theta}$, где $Z$ - статистика критерия, $\hat{\theta}$ - точечная оценка.

Если закон распределения даной статистики известен, в этом случае мы можем построить критическую область.

При условии справедливости нулевой гипотезы $H_0$, статистика $Z$ имеет закон распределения $F_z(x|H_0)$, для определенности будем считать, что статистика $Z$ является непрерывной сл. вел., а значит существует плотность распределения $f_z(\cdot|H_0)$

Рассмотрим, как выглядит критическая область в зависимости от альтернативной гипотезы.

**Правостороний критерий** $H_0: \theta > \theta_0$

Для того, чтобы найти границы критических областей, нам необходимо найти критические значения статистики $Z$:

$F_z(z_k|H_0) = 1 - \alpha$, где $z_k$ - критическая точка, являются квантилями соответствующих порядков ($1 - \alpha$), т.е. $z_k = z_{1 - \alpha}$

$p_{value} = 1 - F_z(z_в|H_0)$, где $z_в$ - выборочное значение статистики, значение, которое мы вычислили по предложенной выборке.

**Левосторонний критерий** $H_0: \theta < \theta_0$

$F_z(z_k|H_0) = \alpha$

$p_{value} = F_z(z_в|H_0)$

**Двухсторонний критерий** $H_0: \theta \neq \theta_0$

$F_z(z^1_k|H_0) = \frac{\alpha}{2}$

$F_z(z^2_k|H_0) = 1 - \frac{\alpha}{2}$

$p_{value} = 2 min (F_z(z_в|H_0), 1 - F_z(z_в|H_0))$

#### Лемма Неймана-Пирсона

Рассмотрим два случая.

**Простая гипотеза**

$H_0: \theta = \theta_0$

$H_1: \theta = \theta_1$

Рассмотрим статистику $\Lambda(x) = \frac{L_x(\theta_1)}{L_x(\theta_0)}$

Если $\exists \:\: c_{\alpha}$ т.ч. $P_{\theta_0}(\Lambda(x) > c_{\alpha}) = \alpha$, то $S = \left\{ \Lambda(x) > с_{\alpha} \right\}$ - критерий уравня значимости, который **имеет максимальную мощность**!

**Сложная гипотеза**

$H_0: \theta = \theta_0$

$H_1: \theta > \theta_0$

Критерий $S$ уровня значимости $\alpha$ называется равносильно наиболее мощным критерием (РНМК), если $\forall$ критерия $R$ ур. знач. $\alpha$ выполнено:

$\gamma_S \geq \gamma_R$ мощность критерия $S$ больше критерия $R$.

Пусть при $\theta_1 > \theta_2$ отношение правдоподобия представимо в виде 

$\frac{L_X(\theta_1)}{L_X(\theta_2)} = f_{\theta_1, \theta_2}(Z)$, где $Z$ статистика.

$f_{\theta_1, \theta_2}(z_в)$ возврастает по $z_в$

Тогда критерий $S = \left\{ Z > с_{\alpha} \right\}$ - РНМК ур. знач. $\alpha$ для 

$H_0: \theta = \theta_0$

$H_1: \theta > \theta_0$

где $с_{\alpha}$ подбирается из условия:

$P_{\theta_0}(Z > с_{\alpha}) = \alpha$

##### Пример. $Exp(\theta)$. Построение наиболее мощного критерия

Пусть $X = (x_1, ..., x_n) \sim Exp(\theta)$.

$H_0: \theta = \theta_0$

$H_1: \theta > \theta_0$

Расписываем отношение правдоподобия при $\theta_1 > \theta_2$

$\frac{L_X(\theta_1)}{L_X(\theta_2)} = \frac{\theta^n_1 e^{- \theta_1 \sum \limits x_i}}{\theta^n_2 e^{- \theta_2 \sum \limits x_i}} = (\frac{\theta_1}{\theta_2})^n e^{(\theta_2 - \theta_1) \sum x_i}$,

т.к. $(\theta_2 - \theta_1) < 0$, то ф-я убывает по $Z = \sum x_i$.

Тогда берем $S = \left\{ \sum x_i < с_{\alpha} \right\}$, где $с_{\alpha}$ выбираем из условия

$P_{\theta_0}(\sum x_i < с_{\alpha}) = \alpha$

Мы знаем, что $\sum x_i \sim Г(\theta, n)$ имеет распределение гамма. Поэтому

$c_{\alpha}$ - $\alpha$ квантиль $Г(\theta_0, n)$

Как вычислить $c_{\alpha}$ в python.

$c_{\alpha} = stats.gamma(a=n, scale = \frac{1}{\theta_0}).ppf(\alpha)$

Мощность

$\gamma(\theta) = stats.gamma(a=n, scale = \frac{1}{\theta}).cdf(c_{\alpha})$


**Вывод:** используя лемму Неймана-Пирсона, мы нашли наиболее мощный (имеющий меньшую вероятность ошибки II рода) критерий (критическую область), на которой отвергаем гипотезу $H_0$, с нужным уровнем значимости $\alpha$

##### Пример. $N(a, \sigma)$. Построение наиболее мощного критерия для мат. ож. $a$

Пусть $\xi \sim N(a, \sigma)$, $a$ - неизвестно, $\sigma^2$ - известно.

$H_0: a = a_0$

$H_0: a = a_1 > a_0$

Проверяем 2 простые гипотезы, т.к. закон распр. ген. совокупности зависит от 2 параметров $a$ и $\sigma^2$, мы предположили, что $\sigma^2$ известно, а гипотезы $H_0$ и $H_1$ нам задают конкретные значения $a$, то мы однозначно определяем распределение, с которым будем работать, поэтому можно применить критерий Неймана-Пирсона.

Применим критерий Неймана-Пирсона

$ L_0(X_{[n]}) = \prod p_0(x_i) = \prod \frac{1}{ \sqrt{2\pi \sigma} } e^{ - \frac{(x_i - a_0)^2}{2\sigma^2} } = \frac{1}{ (2\pi)^{\frac{n}{2}} \sigma^n } e^{- \frac{\sum (x_i - a_0)^2 } {2\sigma^2} }$

$ L_1(X_{[n]}) = \prod p_1(x_i) = \prod \frac{1}{ \sqrt{2\pi \sigma} } e^{ - \frac{(x_i - a_1)^2}{2\sigma^2} } = \frac{1}{ (2\pi)^{\frac{n}{2}} \sigma^n } e^{- \frac{\sum (x_i - a_1)^2 } {2\sigma^2} }$

После сокращения получим следующее выражение. Рассмотрим отношение при $a_1 > a_0$:

$\frac{L_1(X_{[n]})}{L_0(X_{[n]})} = e^{ \frac{1}{2\sigma^2} (2 (a_1 - a_0)n \overline X - n(a^2_1 - a^2_0))) }$

```
n = 50
xs = np.random.normal(loc=0, scale = 1, size=n)
D = 2
a_0 = 1
a_1 = 2

y = np.exp( ((((2 * (a_1 - a_0)) * n * xs) - (n * (a_1 ** 2 - a_0 ** 2)))) / (2 * D) )

plt.plot(xs, y, 'go');
```

Видим, что ф-я возврастает по $\overline X$, тогда берем $S = \left\{ \overline X > с_{\alpha} \right\}$, где $с_{\alpha}$ выбираем из условия

$P_{a_0}(\overline X > с_{\alpha}) = \alpha_0$, где

$\alpha_0$ - вероятность ошибки I рода.

Заметим, что $\xi \sim N(a_0, \sigma)$, тогда $\overline X \sim N(a_0, \frac{\sigma}{\sqrt{n}})$, тогда $с_{\alpha}$ - квантиль нормального распределения из $N(a_0, \frac{\sigma}{\sqrt{n}})$.

Вычислим $c_{\alpha}$ аналитически из условия

$P_{a_0}(\overline X > с_{\alpha}) = \alpha_0$

Из обеих частей неравенства вычтем $a_0$ и поделим на $\frac{\sigma}{\sqrt{n}}$, в этом случае мы получим такую же вероятность, как исходная.

$P_{a_0}(\frac{\overline X - a_0}{\frac{\sigma}{\sqrt{n}}} > \frac{с_{\alpha} - a_0}{\frac{\sigma}{\sqrt{n}}}) $

Заметим, что $\frac{\overline X - a_0}{\frac{\sigma}{\sqrt{n}}} \sim N(0, 1)$, тогда

$P_{a_0}(\frac{\overline X - a_0}{\frac{\sigma}{\sqrt{n}}} > \frac{с_{\alpha} - a_0}{\frac{\sigma}{\sqrt{n}}}) = 1 - Ф(\frac{с_{\alpha} - a_0}{\frac{\sigma}{\sqrt{n}}})$

Поэтому $c_{\alpha}$ мы находим из уравнения

$Ф(\frac{с_{\alpha} - a_0}{\frac{\sigma}{\sqrt{n}}}) = 1 - \alpha_0$  

$с_{\alpha} = a_0 + u_{1-\alpha_0} \frac{\sigma}{\sqrt{n}}$, где

$u_{1-\alpha_0}$ - квантиль стандартного нормального распределения.

**Вывод:**

Мы построили оптимальный критерий $S = \left\{ \overline X > с_{\alpha} \right\}$, который имеет наибольшую мощность. Статистикой критерия является $Z = \overline X$ и критическая область $V_k = (c_{\alpha}, + \infty)$

##### Пример. $N(a, \sigma)$. Построение наиболее мощного критерия для дисперсии $\sigma$

Пусть $\xi \sim N(a, \sigma)$, $a$ - известно, $\sigma^2$ - неизвестно.

$H_0: \sigma^2 = \sigma^2_0$

$H_0: \sigma^2 = \sigma^2_1 > \sigma^2_0$

Проверяем 2 простые гипотезы, т.к. закон распр. ген. совокупности зависит от 2 параметров $a$ и $\sigma^2$, мы предположили, что $a$ известно, а гипотезы $H_0$ и $H_1$ нам задают конкретные значения $sigma^2$, то мы однозначно определяем распределение, с которым будем работать, поэтому можно применить критерий Неймана-Пирсона.

Применим критерий Неймана-Пирсона

$ L_0(X_{[n]}) = \prod p_0(x_i) = \prod \frac{1}{ \sqrt{2\pi \sigma_0} } e^{ - \frac{(x_i - a)^2}{2\sigma^2_0} } = \frac{1}{ (2\pi)^{\frac{n}{2}} \sigma^n_0 } e^{- \frac{\sum (x_i - a)^2 } {2\sigma^2_0} }$

$ L_1(X_{[n]}) = \prod p_1(x_i) = \prod \frac{1}{ \sqrt{2\pi \sigma_1} } e^{ - \frac{(x_i - a)^2}{2\sigma^2_1} } = \frac{1}{ (2\pi)^{\frac{n}{2}} \sigma^n_1 } e^{- \frac{\sum (x_i - a)^2 } {2\sigma^2_1} }$

После сокращения получим следующее выражение

$\frac{L_1(X_{[n]})}{L_0(X_{[n]})} = \big ( \frac{\sigma_0}{\sigma_1} \big )^n e^{ \big ( \frac{1}{2\sigma^2_0} - \frac{1}{2\sigma^2_1} \big ) \cdot \sum(x_i - a)^2 }$

т.к. по условию $\sigma^2_1 > \sigma^2_0$, то выражение будет

$\frac{L_1}{L_0} > 0 $ и будет расти по $\sum(x_i - a)^2$

Тогда берем критерий $S = \left\{ \sum(x_i - a)^2 > с_{\alpha} \right\}$, где $с_{\alpha}$ выбираем из условия

$P_{\sigma^2_0}(\sum(x_i - a)^2 > с_{\alpha}) = \alpha_0$, где

$\alpha_0$ - вероятность ошибки I рода.

Перепишем вероятность, которая будет равнозначна исходной

$P_{\sigma^2_0}(\sum(x_i - a)^2 > с_{\alpha}) = P_{\sigma^2_0}(\frac{\sum(x_i - a)^2}{\sigma^2_0} > \frac{с_{\alpha}}{\sigma^2_0})$

Данная статистика $\frac{\sum(x_i - a)^2}{\sigma^2_0} \sim \chi^2_n$, с $n$ степенями свободы, т.к. известно мат. ож. $a$. Поэтому

$P_{\sigma^2_0}(\sum(x_i - a)^2 > с_{\alpha}) = P_{\sigma^2_0}(\frac{\sum(x_i - a)^2}{\sigma^2_0} > \frac{с_{\alpha}}{\sigma^2_0}) = 1 - F_{\chi^2_n}(\frac{с_{\alpha}}{\sigma^2_0})$

Мы находим $c_{\alpha}$ из следующего уравнения

$\frac{c_{\alpha}}{\sigma^2_0} = \chi^2_{1-\alpha, n}$

$c_{\alpha} = \chi^2_{1-\alpha, n} \cdot \sigma^2_0$, где

$\chi^2_{1-\alpha, n}$ - квантиль хи-квадрат распределения

Т.к. у нас правостороння гиротеза

$H_0: \sigma^2 = \sigma^2_0$

$H_0: \sigma^2 = \sigma^2_1 > \sigma^2_0$

Тогда критическая область будет

$V_k = ( c_{\alpha}, + \infty )$

**Вывод:**

Мы построили оптимальный критерий $S = \left\{ \sum (x_i - a)^2 > с_{\alpha} \right\}$, который имеет наибольшую мощность. Статистикой критерия является $Z = \sum (x_i - a)^2 $ и критическая область $V_k = (c_{\alpha}, + \infty)$

##### Пример. $Bin(n, p)$. Построение наиболее мощного критерия для вероятности $p$

$H_0: p = p_0$

$H_1: p = p_1 > p_0$

$L_1(m) = C^m_n p^m_1 (1 - p_1)^{n-m}$

$L_0(m) = C^m_n p^m_0 (1 - p_0)^{n-m}$

$\frac{L_1(m)}{L_0(m)} = \big ( \frac{p_1}{p_0} \big )^m \big ( \frac{1 - p_1}{1 - p_0} \big )^{n-m}$

Домножим и разделим это выражение на $\big ( \frac{1 - p_1}{1 - p_0} \big )^m$, тогда получим

$\big (\frac{ p_1(1 - p_0)  }{p_0(1 - p_1)}\big )^m \big ( \frac{1 - p_1}{1 - p_0} \big )^n > 0$, т.к. по условию $p_1 > p_0$, это будет эквивалентно, что какая-то величина $m > 0$. Мы видим, что функция растёт по $m$, чем больше успехов тем больше значение ф-и.

Тогда берем критерий $S = \left\{ m > с_{\alpha} \right\}$, где $с_{\alpha}$ выбираем из условия

$P_{p_0}(m > с_{\alpha}) = \alpha_0$, где

$\alpha_0$ - вероятность ошибки I рода.

Если воспользоваться ассимптотическими свойствами оценки параметра $\hat{p}$ биноминального распр., то можно построить приближенный критерий.Рассмотрим вер-ть Ошибки I рода $\alpha$.

$\alpha = P_0(m > c_{\alpha}) + \varepsilon P_0(m = c_{\alpha})$

Из предельной теоремы Муавра-Лапласа следует, что вер-ть $\alpha$ стремится к

$\alpha = P_0(m > c_{\alpha}) + \varepsilon P_0(m = c_{\alpha}) \xrightarrow[n\to\infty]{} 1 - Ф(\frac{c_{\alpha} - np_0}{\sqrt{np_0(1-p_0)}})$

Рассмотрим уравнение

$Ф(\frac{c_{\alpha} - np_0}{\sqrt{np_0(1-p_0)}}) = 1 - \alpha_0$

Из этого ур-я найдём $c_{\alpha} = np_0 + \sqrt{np_0(1-p_0)} u_{1-\alpha_0} $, где

$u_{1-\alpha_0} $ - квантиль стандартного норм. распр.


Возьмем статистику критерия

$Z = m$, где $m$ - кол-во успехов.

Построим критическую область

$V_k = (np_0 + \sqrt{np_0(1-p_0)} u_{1-\alpha_0}, + \infty)$

На практике, часто кол-во успехов не будет +$\infty$, поэтому интервал будет

$V_k = (np_0 + \sqrt{np_0(1-p_0)} u_{1-\alpha_0}, m)$

Если $m \in V_k \Rightarrow H_0$ отклоняем.


**Критическая область для левостороннего критерия**

$V_k = (- \infty, np_0 + \sqrt{np_0(1-p_0)} u_{\alpha_0})$

На практике часто число успехов не может быть отрицательным, тогда граница будет

$V_k = (0, np_0 + \sqrt{np_0(1-p_0)} u_{\alpha_0})$

**Вывод:**

Построенный критерий $S = \left\{ m > с_{\alpha} \right\}$, где $с_{\alpha}$  и построенная критическая область $V_k$ являются приближёнными и следовательно вер-ть Ошибки I рода $\alpha \neq \alpha_0$ не равна конкретному значению $\alpha_0$, а равна лишь приближенно, но чем боьлше будет выборка, чем больше испытаний, тем вероятность Ошибки I рода будут стремится к установленному значению $\alpha \xrightarrow[n\to\infty]{} \alpha_0$.

#### Определение необходимого объема выборки для обеспечения нужной вероятности Ошибки I рода

Продолжим рассматривать случай выше

Пусть $\xi \sim N(a, \sigma)$, $a$ - неизвестно, $\sigma^2$ - известно.

$H_0: a = a_0$

$H_0: a = a_1 > a_0$

$Z = \overline X$

$V_k = (c_{\alpha}, + \infty)$

Предположим, что 

$с_{\alpha} \geq a_0 + u_{1-\alpha_0} \frac{\sigma}{\sqrt{n}}$ в этом случае вер-ть Ошибки I рода $\alpha$, будет не больше, чем некоторая заданная вероятность $\alpha_0$

$\alpha \leq \alpha_0$

Выясним при каком объеме выборки вер-ть Ошибки II рода $\beta$ будет не меньше, чем заданная веростность $\gamma_0$ или 

$\gamma \geq \gamma_0 = 1 - \beta_0$

Вероятность Ошибки II рода 

$\beta = P(\overline X \leq c_{\alpha}|H_1) = P_{a_1}(\frac{\overline X - a_1}{\frac{\sigma}{\sqrt{n}}} \leq \frac{с_{\alpha} - a_1}{\frac{\sigma}{\sqrt{n}}}|H_1)$

Заметим, что $\frac{\overline X - a_1}{\frac{\sigma}{\sqrt{n}}} \sim N(0, 1)$, тогда

$P_{a_1}(\frac{\overline X - a_1}{\frac{\sigma}{\sqrt{n}}} \leq \frac{с_{\alpha} - a_1}{\frac{\sigma}{\sqrt{n}}}|H_1) = Ф(\frac{с_{\alpha} - a_1}{\frac{\sigma}{\sqrt{n}}}) $

Тогда вер-ть Ошибки II рода $\beta$ будет не больше заданного значения $\beta_0$, при условии

$\frac{с_{\alpha} - a_1}{\frac{\sigma}{\sqrt{n}}} \leq u_{\beta_0}$, где

$u_{\beta_0}$ - квантиль стандартного нормального распределения порядка $\beta_0$.

Совместим условия для $c_{\alpha}$ из этого неравенства мы получаем условие для размера выборки

$a_0 + u_{1-\alpha_0} \frac{\sigma}{\sqrt{n}} \leq с_{\alpha} \leq a_1 + u_{\beta_0} \frac{\sigma}{\sqrt{n}}$

Если 

$n \geq \frac{\sigma^2 (u_{1-\alpha_0} - u_{\beta_0})^2}{(a_1 - a_0)^2}$

тогда $\alpha \leq \alpha_0$ и $\beta \leq \beta_0$

Десять разных термометров сопротивления откалибровали по стандартному термометру. При показании стандартного термометра 1000мВ, испытуемые термометры показали:

Найдите p-значение и примите статистическое решение.

**Решение:**

$a$ = 1000

$\alpha_0$ = 0.05

$H_0: a_0 = 1000$

$H_1: a_1 \neq 1000$

Вычислим $p_{value}$, т.к. дисперсия неизвестна, возьмем несмещенную оценку дисперсии

$s^2 = \frac{1}{n-1} \sum\limits_{i=1}^n (x_i - \overline X)^2$

$s = \sqrt{\frac{1}{n-1} \sum\limits_{i=1}^n (x_i - \overline X)^2}$

##### Проверка гипотезы о дисперсии

Пусть $\xi \sim N(a, \sigma)$

$H_0: \sigma^2 = \sigma^2_0$

$H^1_1: \sigma^2 = \sigma^2_1 \neq \sigma^2_0$

$H^3_1: \sigma^2 < \sigma^2_0$

$H^4_1: \sigma^2 \neq \sigma^2_0$

В качестве точечной оценки мат. ож. возьмем выборочное среднее $\overline X$.

Возьмем в качестве статистики критерия

ниже $\chi^2_{xxx, (n-1)}$ - квантиль распределения Стьюдента порядка xxx.

###### $a$ известно, $\sigma^2$ неизвестно

В случае, если мат. ож. $a$ известно, нам не нужно вычислять её оценку. В качестве статистики 

$Z' = \frac{\sum (x_i - a)^2}{\sigma^2_0} \sim \chi^2_{n}$, здесь имеет $n$ степеней свободы, т.к. нет уравнений связи, типо выборочного среднего $\overline X$, которое связывает все $x_i$, мы в качестве $a$ берем само известное значение мат. ож. Интервалы находим так же, как в случае с неизвестным $a$, единственное отличие, мы ищим $\chi^2$ с $n$ степенями свободы!

$H_1: \sigma^2 > \sigma^2_0 \:\:\:\:\:\:\:\:\:\:\:\:\: V_k: (\chi^2_{1 - \alpha, n}, + \infty) \:\: \:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:p_{value} = 1 - F_{\chi^2}(z_в)$

$H_1: \sigma^2 = \sigma^2_1 > \sigma^2_0 \:\: V_k: (\chi^2_{1 - \alpha, n}, + \infty) \:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\: p_{value} = 1 - F_{\chi^2}(z_в)$

$H_1: \sigma^2 < \sigma^2a_0 \:\:\:\:\:\:\:\:\: V_k: (- \infty, \chi^2_{\alpha, n} ) \:\: \:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\:\: p_{value} = F_{\chi^2}(z_в)$

$H_1: \sigma^2 \neq \sigma^2_0 \:\:\:\:\:\:\:\:\:\:\:\:\: V_k: (- \infty, \chi^2_{\frac{\alpha}{2}, n} ) \cup (\chi^2_{1 -\frac{\alpha}{2}, (n)}, + \infty) \:\: p_{value} = 2 \cdot min(F_{\chi^2}(z_в), 1 - F_{\chi^2}(z_в))$

где $F_{\chi^2}$ - ф-я распределения хи-квадрат.

Отдел контроля качества отобрал 25 упаковок и рассчитал несмещённую оценку дисперсии $s^2 = \frac{1}{n-1} \sum (x_i - \overline X)^2$, она составила 13.5.

Есть ли основания считать, что дисперсия объема зубной пасты превышает допустимый предел? Используйте уровень значимости 0.05. Предполагается, что объем зубной пасты в упаковке подчиняется нормальному закону распределения.

**Решение:**

$s^2 = 13.5$

$\alpha = 0.05$

$a = 100$

$n = 25$

$H_0: \sigma^2_0 = 9$

$H_1: \sigma^2_1 > 9$

##### Задача №1. Построение приближ. наиболее мощного критерия для $Bin(n, p)$

Производитель одежды хочет узнать, какие футболки предпочитает целевая группа: однотонные без рисунка или футболки с рисунком. Правда ли, что производителю следует выпускать больше однотонных футболок без рисунка, чем футболок с рисунком, чтобы удовлетворить пожелания покупателей?

Для проверки этой гипотезы  были опрошены 225 человек, 90 высказались в пользу футболок с рисунком, а 135 – в пользу футболок без рисунка.

Будем считать, что "успех" = покупатель предпочитает футболку без рисунка.

Используя приближенный критерий, постройте критическую область и примите статистическое решение при уровне значимости 0.01.

**Решение:**

Рассмотрим гипотезу

$H_0: p_0 = 0.5$

$H_1: p_1 > 0.5$

Т.к. у нас $H_1: p_1 > 0.5$ правосторонний критерий, для построения критической области используем полученную на лекции формулу

$V_k = (np_0 + \sqrt{np_0(1-p_0)} u_{1-\alpha_0}, + \infty)$, где

$u_{1-\alpha_0}$ - квантиль стандартного норм. распр.

Но так как нельзя получить кол-во успехов больше чем было испытаний, то критическая область будет равна

$V_k = (np_0 + \sqrt{np_0(1-p_0)} u_{1-\alpha_0}, m)$

$Z = m = 135$ - кол-во успехов, наша статистика

$n = 225$ 

$\alpha_0 = 0.01$

Найдём квантиль $u_{1-\alpha_0}$


```
alpha = 0.01
u = st.norm.ppf(1 - alpha)

# найдём границу интервала
n = 225
p = 0.5
left_side = (n * p) + ( np.sqrt( n * p * (1 - p) ) * u)
left_side
```

$V_k = (130, 225)$

$ m \in V_k \Rightarrow H_0$ отклоняется.

**Ответ:**

$V_k = (130, 225)$

$ H_0$ отклоняется

Согласно этим данным только 60 из 747 человек скончались в течение трехмесячного периода, предшествующего их дни рождения. Подтверждают ли эти данные предположение?

$\alpha = 0.02$

$Z = m = 60$ наша статистика, кол-во успехов, человек скончался в течение трехмесячного периода, предшествующего его дня рождения.
