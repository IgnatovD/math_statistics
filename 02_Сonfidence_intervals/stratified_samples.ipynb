#### Теория. Мат. ож. и дисп. смеси из распр.

Смесь нескольких распределений

Пусть $ \xi \sim F(x) = W_1F_1(x) + ... + W_lF_l(x)$

$ W_k $ - доля страты $ k = \overline {1,L} $

Пусть 

$ \mu = E \xi $

$ \sigma^2 = D \xi $

$ \mu_k $ - мат. ож. страты $k$

$ \sigma^2_k $  - дисперсия страты $k$

Мат. ожидание

$ \mu = E \xi = \int\limits_{-\infty}^{+\infty} x dF(x) = \int\limits_{-\infty}^{+\infty} xd(W_1F_1(x) + ... + W_LF_L(x) ) = \sum\limits_{k=1}^L W_k \int\limits_{-\infty}^{+\infty} xdF_k(x) = \sum\limits_{k=1}^L W_k E \xi_k = \sum\limits_{k=1}^L W_k \mu_k $

$ \mu = \sum\limits_{k=1}^L W_k \mu_k $

Дисперсия

$ D \xi =  \int\limits_{-\infty}^{+\infty} (x - E \xi)^2 dF(x) = \int\limits_{-\infty}^{+\infty} (x - E \xi)^2 d ( W_1F_1(x) + ... + W_LF_L(x) ) = \sum\limits_{k=1}^L W_k \int\limits_{-\infty}^{+\infty} (x - E \xi)^2 dF_k(x) = $ 
в рамках каждого k-го слагаемого мы добавим и вычтем $E\xi$ в $ (x - E \xi)^2 $

$ \sum\limits_{k=1}^L W_k ( \int\limits_{-\infty}^{+\infty} (x - E \xi)^2 dF_k(x)  + \int\limits_{-\infty}^{+\infty} (E \xi_k - E \xi)^2 dF_k(x) ) = \sum\limits_{k=1}^L W_k (\sigma^2_k + (\mu_k - \mu)^2)$

Итого получили 

$ D \xi = \sum\limits_{k=1}^L W_k (\sigma^2_k + (\mu_k - \mu)^2)  $

#### Теория. Оценка мат. ож. и дисп.

В рамках страты $k$ возьмем выборку объема $n_k$ $(X_1k, ..., X_{n_k}k) $, $ k = \overline {1, L}$

Выборочное среднее

$ \overline {X_k} = \frac{1}{n_k} \sum\limits_{i=1}^{n_k} x_{ik} $, где

$n_k$ - кол-во эл-ов в $k$-ой выборке

$ x_{ik} $ - $i$-й эл-т $k$-ой выборки

Выборочная дисперсия

$ D^*_k = \frac{1}{n_k} \sum\limits_{i=1}^{n_k} ( x_{ik} - \overline x_k)^2 $, где

$ \overline x_k $ - среднее значение $k$-ой выборки

##### Оценка мат. ож. смеси



$ \overline X_s = \sum\limits_{k=1}^L W_k \overline x_k $

Найдём мат. ож. оценки

$E \overline X_s = \sum\limits_{k=1}^L W_k E \overline x_k = |т.к. \overline x_k \: является \: не \: смещенной \: оценкой \: мат. \: ож., \: то \: справедлив \: переход | \sum\limits_{k=1}^L W_k \mu_k = \mu$, где

$\mu$ - мат. ож. генеральной совокупности

**Свойство:** не смещенная оценка

Найдём дисперсию оценки

$ D(\overline X_s) = \sum\limits_{k=1}^L W^2_k D(\overline x_k) = \sum\limits_{k=1}^L W^2_k \frac{\sigma^2_k}{n_k} $, где

$ \sigma^2_k $ - дисперсия $k$-й страты

##### Оценка дисперсии смеси

Дисперсия генеральной совокупности равна

$ D \xi = \sum\limits_{k=1}^L W_k \sigma^2_k + \sum\limits_{k=1}^L W_k (\mu - \mu_k)^2) $

Оценка дисперсии смеси

$ D^*_s = \sum\limits_{k=1}^L W_k D^*_k + \sum\limits_{k=1}^L W_k (\sum\limits_{i=1}^L W_i \overline x_i - \overline x_k )^2 $, где

$ D^*_k $ - выборочная дисперсия $k$-й страты

$ \overline x_i $ - среднее значение $i$-й страты

$ \sum\limits_{i=1}^L W_i \overline x_i = \overline X_s $ - оценка мат. ож. смеси распределений

$ \overline x_k $ - среднее значение $k$-й страты

**Свойство:** смещенная оценка, но ассимптотически не смещенная


#### Задача №2. Оценка мат. ож. и дисп.

Предположим, что необходимо оценить число свободных мест в больницах. Все больницы разбили на 2 страты по числу коек-мест. Доля страты 1 составила 40%, страты 2 - 60%.

Из каждой страты были получены выборки числа свободных мест объемом 10 каждая:

страта 1: 282, 226, 188, 327, 344, 304, 414, 224, 335, 270

страта 2: 417,  851,  742, 1217, 1160,  993,  864,  852, 1286,  988

Найдите оценку среднего числа свободных мест в больницах $(\overline X_s)$, а также оценку дисперсии числа свободных мест $(D^*_s)$


```
w = np.array([0.4, 0.6])
s1 = np.array([282, 226, 188, 327, 344, 304, 414, 224, 335, 270])
s2 = np.array([417, 851, 742, 1217, 1160, 993, 864, 852, 1286, 988])
```

Оценка мат. ож.

$ \overline X_s = \sum\limits_{k=1}^L W_k \overline x_k $

```
# вектор средних значений по стратам
ex = np.array([s1.mean(), s2.mean()])
# вектор оценок дисперсий
ed = np.array([s1.var(), s2.var()])
# оценка мат. ож. смешанных распределений
xs = w @ ex
```

Оценка дисперсии смеси

$ D^*_s = \sum\limits_{k=1}^L W_k D^*_k + \sum\limits_{k=1}^L W_k (\sum\limits_{i=1}^L W_i \overline x_i - \overline x_k )^2 $

оценка дисперсии страты

$ D^*_k = \frac{1}{n_k} \sum\limits_{i=1}^{n_k} (x_i - \overline x_k)^2 $

```
# оценка смешанной дисперсии
ds = w @ ed + w @ (xs.repeat(2) - ex)**2
print('Оценка мат. ож. смеси:', round(xs, 1), 'Оценка дисперсии смеси:', round(ds, 1))

# пример того, как много кода писать, если не использовать векторы, ниже код второго слагаемого
#w[0] * (xs - ex[0])**2 + w[1] * (xs - ex[1])**2
```

Решение на python

```
a = [282, 226, 188, 327, 344, 304, 414, 224, 335, 270]
b = [417, 851, 742, 1217, 1160, 993, 864, 852, 1286, 988]

m = 0.4 * np.mean(a) + 0.6 * np.mean(b)
d = 0.4 * np.std(a)**2 + 0.6 * np.std(b)**2 + 0.4 * (np.mean(a) - m)**2 + 0.6 * (np.mean(b) - m)**2
print(f'{m:.1f}, {d:.1f}')
```

#### Теория. Какими выбрать выборки из страт?

Теорема

Оптимальное сэмплирование Неймана

Объемы выборок $ n_1, ..., n_L $ такие, что 

$ n_k = \tilde n_k = n \frac{W_k \sigma_k}{\sum\limits_{i=1}^L W_i \sigma_i } $ **минимизируют** дисперсию $D(\overline X_s)$, где

$ k = \overline {1, k} $, $n = n_1 + ... + n_L$

$ n $ - кол-во в генеральной совокупности

$ n_k $ - объем страты

$ \tilde n_k $ - объем взятой выборки согласно условиям теоремы

$ \sigma_k = \sqrt{D^*_k} $ - стандартное отклонение

Пропорциональное сэмплирование

$ n_k = nW_k,  k = \overline {1, k} $, $n = n_1 + ... + n_L$

#### Задача №3. Оптимальное сэмплирование Неймана

Генеральная совокупность представляет собой смесь двух распределений: первое составляет 25% исходной генеральной совокупности, имеет дисперсию 4, второе - 75%,  дисперсию 6.25. Какими следует выбрать объемы выборок из данных страт так, чтобы оценка мат.ожидания исходной генеральной совокупности имела наименьшую дисперсию, если общий объем выборки должен равняться 100?

**Решение:**



```
w = np.array([0.25, 0.75])
var = np.array([4, 6.25])
sigma = np.sqrt(var)
n = 100
```

```
n_1 = n * (w[0] * sigma[0]) / (w @ sigma)
n_2 = n * (w[1] * sigma[1]) / (w @ sigma)
print(f'{n_1:.0f}, {n_2:.0f}')
```

**Ответ:**

21, 79
