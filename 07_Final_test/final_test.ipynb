#### 8.1 задача

Какие из следующих утверждений верные?

* Точечная оценка является случайной величиной - верно.

Точечная оценка является случайной величиной, т.к. она является статистикой (зависим только от выборки). Т.к. результаты опытов случайны, любая статистика является случайной величиной. [объяснение](https://stepik.org/lesson/39623/step/2?unit=45734).

* С уменьшением уровня значимости, мощность увеличивается - неверно.

С уменьшением уровня значимости (ошибки I рода), увеличивается вероятность ошибки II рода ($\beta$). Т.к. мощность критерия равна $1 - \beta$, соответственно будет увеличиваться $\beta$ и уменьшаться можность критерия, т.е. при уменьшении вер-ти ошибки I рода, будет уменьшаться мощность критерия. [объяснение](http://www.machinelearning.ru/wiki/index.php?title=Уровень_значимости)

* Критическая область является дополнением области принятия гипотеза - верно.

Множество всех возможных значений статистики $Z$ является $V \cup V_k$, где $V$ область принятия гипотезы, $V_k$ критическая область.

* С уменьшением объема выборки вероятность ошибки I рода увеличивается - неверно.

$\alpha = P(Z \in V_k |H_0) = P(X_{[n]} \in S |H_0)$, где

$\alpha$ - вер-ть ошибки I рода

$S$ область всех возможных значений выборки $X_{[n]}$

$Z$ статистический критерий

$|H_0$ при условии, что $H_0$ верна.

Как видно из формул, вероятность ошибки I рода не зависит от объема выборки.

* Выборочное среднее является смещенной оценкой мат.ожидания - неверно.

$E(\overline X) = E(\frac{1}{n} \sum X_i) = \frac{1}{n} E( \sum X_i) = \frac{1}{n} \sum E(X_i) = \frac{1}{n} \sum a = a $, где $a$ - мат. ож. ген. совокупности.

т.к. $X_i$ имеют одинаковый закон распределения, как и у генеральной совокупности, значит они имеют одинаковое мат. ожидание, как и у генеральной совокупности.

Т.к. мат. ож. оценки параметра равно теоритическому мат. ожиданию ($E \hat{\theta} = \theta$), отсюда следует, что выборочное среднее это несмещенная оценка мат. ож.

*  С ростом объема выборки мощность критерия увеличивается - верно

* Уровень значимости статистического критерия равен вероятности того, что нулевая гипотеза верна - неверно.

* С ростом объема выборки дисперсия выборочного среднего уменьшается - верно.

$D (\overline X) = D \bigg (\frac{\sum \limits_{i=1}^n X_i}{n} \bigg ) $

вынесем константу $\frac{1}{n}$ из под знака дисперсии, воспользуемся свойством дисперсии $D(kX) = k^2 D(X)$

$D (\overline X) = D \bigg (\frac{\sum \limits_{i=1}^n X_i}{n} \bigg ) = \frac{1}{n^2} D(\sum \limits_{i=1}^n X_i) $

воспользуемся свойством $D(X \pm Y) = D(X) + D(Y)$

$D (\overline X) = D \bigg (\frac{\sum \limits_{i=1}^n X_i}{n} \bigg ) = \frac{1}{n^2} D(\sum \limits_{i=1}^n X_i) = \frac{1}{n^2} \sum \limits_{i=1}^n D(X_i)$

т.к. $X_i$ имеют одинаковый закон распределения, как и у генеральной совокупности, значит они имеют одинаковую дисперсию, как и у генеральной совокупности.

$D (\overline X) = D \bigg (\frac{\sum \limits_{i=1}^n X_i}{n} \bigg ) = \frac{1}{n^2} D(\sum \limits_{i=1}^n X_i) = \frac{1}{n^2} \sum \limits_{i=1}^n D(X_i) = \frac{n \cdot D(X_{1})}{n^2} = \frac{D(X_1)}{n}$

$\sigma^2_{\overline X} (\overline X) = \sigma^2 \bigg (\frac{\sum \limits_{i=1}^n X_i}{n} \bigg ) = \frac{1}{n^2} \sigma^2 (\sum \limits_{i=1}^n X_i) = \frac{1}{n^2} \sum \limits_{i=1}^n \sigma^2_{x_i} = \frac{n \cdot \sigma^2}{n^2} = \frac{\sigma^2}{n}$

С ростом $n$ дисперсия будет стремится к нулю.

#### 8.4 задача

Случайная величина $\xi$ имеет следующее вероятностное распределение

$P(\xi = 0) = \theta  \cdot \frac{2 }{3}$

$P(\xi = 1) = \theta  \cdot \frac{1 }{3}$

$P(\xi = 2) = (1 - \theta)  \cdot \frac{2 }{3}$

$P(\xi = 3) = (1 - \theta)  \cdot \frac{1 }{3}$, где

$\theta \in (0, 1)$

По выборке 3, 0, 2, 1, 3, 2, 1, 0, 2, 1

найдите с  помощью метода моментов оценку параметра $\theta$

**Решение:**

По методу моментов, нам необходимо приравнять теоритическое мат. ож. ф-и распределения к его выборочному аналогу, т.е. выборочному среднему.







Найдём теоритическое мат. ож.

$E(X) = \sum \limits_{i=1}^n x_i p_i $, где

$x_i$ значение сл. величины

$p_i$ вер-ть значения сл. вел.

$E(\xi) = \sum \limits_{i=1}^n x_i p_i  = 0 \cdot ( \theta  \cdot \frac{2 }{3} ) + 1 \cdot ( \theta  \cdot \frac{1 }{3}) + 2 \cdot ( (1 - \theta)  \cdot \frac{2 }{3}) + 3 \cdot ( (1 - \theta)  \cdot \frac{1 }{3}) = 0 + \frac{\theta}{3} + \frac{(1 - \theta) \cdot 4}{3} + \frac{(1 - \theta) \cdot 3}{3} = \frac{\theta + (1 - \theta) \cdot 4 + (1 - \theta) \cdot 3}{3} = \frac{\theta + 4 - 4\theta + 3 - 3\theta}{3} = \frac{7 - 6\theta}{3} = \frac{7}{3} - \frac{6\theta}{3} = \frac{7}{3} - 2 \theta$

Получили, теоритическое мат. ож. равно

$E(\xi) = \frac{7}{3} - 2 \theta$

Приравняем теоритическое мат. ож. к его выборочному аналогу, т.е. выборочному среднему

$\frac{7}{3} - 2 \theta = \frac{1}{n} \sum x_i$

Отсюда выразим искомый параметр $\theta$

$\frac{7}{3} - \frac{1}{n} \sum x_i  = 2 \theta $

$ \frac{\frac{7}{3} - \frac{1}{n} \sum x_i }{2}  = \theta $

$\theta = \frac{\frac{7}{3} - \overline X}{2}$

```
X = np.array([3, 0, 2, 1, 3, 2, 1, 0, 2, 1])

Xmean = X.mean()

theta = ((7/3) - Xmean) / 2
theta
```

**Ответ:**

$\theta = 0.416$

#### 8.5 задача

Случайная величина $\xi$ имеет следующее вероятностное распределение

$P(\xi = 0) = \theta  \cdot \frac{2 }{3}$

$P(\xi = 1) = \theta  \cdot \frac{1 }{3}$

$P(\xi = 2) = (1 - \theta)  \cdot \frac{2 }{3}$

$P(\xi = 3) = (1 - \theta)  \cdot \frac{1 }{3}$, где

$\theta \in (0, 1)$

По выборке 3, 0, 2, 1, 3, 2, 1, 0, 2, 1

найдите с  помощью метода моментов оценку параметра $\theta$



**Решение:**

Необходимо составить функцию правдоподобия, т.к. у нас распределение дискретно, то ф-я правдоподобия будет 

$ L_X = \prod\limits_{i = 1}^n p_{\theta}(x_i)  $, где

$p_{\theta}(x) = P(\xi = x)$


$L_X = L_X = \prod\limits_{i = 1}^n p_{\theta}(x_i) = \theta  \cdot \frac{2 }{3} \cdot \theta  \cdot \frac{1 }{3} \cdot (1 - \theta)  \cdot \frac{2 }{3} \cdot (1 - \theta)  \cdot \frac{1 }{3} = (\theta)^2 \cdot (1-\theta)^2 \cdot (\frac{1 }{3})^2 \cdot (\frac{2 }{3})^2 = (\theta)^2 \cdot (1-\theta)^2 \cdot \frac{4}{81}$

Прологарифмируем ф-ю правдоподобия

$l_X = ln(L_X) = ln \bigg ( (\theta)^2 \cdot (1-\theta)^2 \cdot \frac{4}{81} \bigg ) = 2 ln\theta + 2 ln(1 - \theta) + ln(4) - ln(81) \rightarrow \underset{\theta}{\operatorname{max}}$

Найдём производную по $\theta$

$\frac{\delta l_X}{\delta \theta} = \frac{2}{\theta} - \frac{2}{1 - \theta}$

Запишем уравнение правдоподобия и выразим искомый параметр

$\frac{2}{\theta} - \frac{2}{1 - \theta} = 0$

$\frac{2}{\theta} = \frac{2}{1 - \theta}$

$\frac{2 \cdot (1 - \theta)}{\theta} = 2$

$\frac{2  - 2\theta}{\theta} = 2$

$\frac{2}{\theta} - \frac{2\theta}{\theta} = 2$

$\frac{2}{\theta} - 2 = 2$

$\frac{2}{\theta} = 4$

$\frac{1}{\theta} = 2$

$\theta = \frac{1}{2}$

**Ответ:**

$\theta = \frac{1}{2}$



#### 8.6 задача

После введения новых штрафов за превышение скорости замерили скорость движения автомобилей по КАД (км/ч):

128,107,107,92,118,114,117,109,115,106,109,106,109,120,112,128,114,119,110,130

Правда ли, что в среднем скорость движения по КАД стала выше 110 км/ч? Проверьте данную гипотезу, предполагая , что скорость движения подчиняется нормальному закону распределения. Уровень значимости 0.03

Найдите p-значение и примите статистическое решение

**Решение:**

Проверим гипотезу

$H_0: a = 110$

$H_1: a > 110$



Решим используя $p_{value}$

Т.к. предполагаем, что выборка взята из нормального распределения и нам неизвестно мат. ож. и дисперсия, то для проверки гипотезы будем использовать критерий

$Z = \frac{\overline X - a_0}{\frac{s}{\sqrt{n}}} \sim T_{n-1}$, где 

$s^2 = \frac{1}{n-1} \sum\limits_{i=1}^n (x_i - \overline X)^2$ - несмещенная оценка дисперсии.

$S = \sqrt{S^2}$

Проверим гипотезу с помощью $p_{value}$

$p_{value} = 1 - F_{Z}(z_в) = 1 - T_{n-1}(z_в)$ 

```
X = np.array([128,107,107,92,118,114,117,109,115,106,109,106,109,120,112,128,114,119,110,130])
n = len(X)
# ddof=1 несмещенная оценка дисперсии
Zv = (X.mean() - 110) / ( np.sqrt(X.var(ddof=1)) / np.sqrt(n))
Zv
```

```
st.t.sf(Zv, df = n - 1)
```

$ \alpha = 0.03$

$p_{value} > \alpha \Rightarrow H_0$ не отвергается. Значит средняя скорость равна 110 км. ч.

Решим, через построение критической области.

$V_k = (t_{1 - \alpha, n - 1}, + \infty)$

```
alpha = 0.03
st.t.ppf(1 - alpha, df = n - 1)
```

$Z_в = 1.74$

$V_k = (2, + \infty)$

$Z_в \notin V_k \Rightarrow H_0$ не отвергается. Значит средняя скорость равна 110 км. ч.

**Ответ:**

$p_{value} = 0.04$

$H_0$ не отвергается.

#### 8.7 задача

По данным журнала Forbes  доходы топ 9 самых богатых людей мира 2016 на начало 2015 и 2016  годов были следующими

2015 [79.2, 65.0, 72.7, 77.1, 34.8, 33.4, 54.3, 35.5, 42.9]

2016 [75, 67, 60.8, 50, 45.2, 44.6, 43.6, 40, 39.6]

Верно ли то, что по сравнению с началом 2015 года  доходы уменьшились?
Воспользуйтесь критерием знаков для проверки данной гипотезы. Уровень значимости 0.05

**Решение:**

Получена двумерная выборка $(\eta, \xi)$, $(X, Y)$.

Требуется проверить гипотезу о равенстве ф-й распределений, т.е. мы проверяем, есть ли изменения или нет.

$H_0: F(X) = G(Y)$ для всех $x \in \mathbb {R}$

$H_1: F(X) \leq G(Y)$ для всех $x \in \mathbb {R}$

Введем новую сл. величину $\zeta = \eta - \xi$, тогда $z_i = x_i - y_i$

Поэтому давайте вместо гипотезы $H_0: F(x) = G(x)$ будем рассматривать гипотезу 

$H_0: P(z_i < 0) = P(z_i > 0) = \frac{1}{2}$

$H_1: P(z_i < 0) > P(z_i > 0)$

Определим успех как $z_i < 0$, тогда выборка $Z = (z_1, ..., z_n)$ это последовательность испытаний по схеме Бернулли и в случае справедливости $H_0$ вер-ть появления успеха равна $\frac{1}{2}$

Возьмем статистический критерий

$L = \sum I(z_i < 0) \sim Bin(n, \frac{1}{2})$ в случае, когда справедлива $H_0$

Статистика принимает значения из интервала

$L \in [0, n]$



Критическая область

$H_1: F(X) \leq G(Y)$

$V_k: [0, c_2)$

```
X = np.array([79.2, 65.0, 72.7, 77.1, 34.8, 33.4, 54.3, 35.5, 42.9])
Y = np.array([75, 67, 60.8, 50, 45.2, 44.6, 43.6, 40, 39.6])
Z = X - Y

Zv = np.sum([1 for z in Z if z < 0])
```

```
alpha = 0.05
n = len(Z)
p0 = 0.5
c_alpha = st.binom.ppf(alpha, n, p0)
c_alpha
```

$V_k = [0, 2)$

$Z_в = 4$

$Z_в \notin V_k \Rightarrow H_0$ не отвергается. Значит доходы остались такими же.

**Ответ:**

$Z_в = 4$

$V_k = [0, 2)$

$H_0$ не отвергается.

#### 8.8 задача

Курение считают основным фактором, предрасполагающим к хроническим заболеваниям легких. Что касается пассивного курения, оно таким фактором обычно не считается. Дж. Уайт и Г. Фреб усомнились в безвредности пассивного курения и исследовали проходимость дыхательных путей у некурящих, пассивных и активных курильщиков (J. White, H. Froeb. Small-airways dysfunction in nonsmokers chronically exposed to tobacco smoke. N. Engl. J. Med., 302:720—723, 1980). Для характеристики состояния дыхательных путей взяли один из показателей функции внешнего дыхания — максимальную объемную скорость середины выдоха, которую измеряли во время профилактического осмотра сотрудников Калифорнийского университета в Сан-Диего. Уменьшение этого показателя — признак нарушения проходимости дыхательных путей. Данные обследования представлены в таблице.

```
df = pd.DataFrame({'Группа': ['некурящие, в помещении не курят', 'некурящие, в помещении курят', 'курящие, выкуривают мало сигарет', 'курящие, выкуривают среднее число сигарет', 'курящие, выкуривают много сигарет'], 'число обследованных': [200, 200, 200, 200, 200], 'внутригрупповое среднее': [3.17, 2.72, 2.63, 2.29, 2.12], 'внутригрупповая выборочная дисперсия': [0.5476, 0.5041, 0.5329, 0.49, 0.5184]})
df
```

Можно ли считать максимальную объемную скорость середины выдоха одинаковой во всех группах?

Предполагаем, что выборки получены из нормально распределенных генеральных совокупностей с равными дисперсиями. Уровень значимости 0.05.


**Решение:**

Проверим гипотезу о равенстве мат. ож. генеральных совокупностей, если эта гипотеза выполняется, то все ген. совокупности имеют один и тот же закон распределения и следовательно фактор не влияет на зависимую переменную.

$H_0: a_1 = a_2 = ... = a_k$

$H_1: \exists t,s: a_t \neq a_s$

Вычислим статистики

$Q_1 = \sum \limits_{j=1}^k \sum \limits_{i=1}^{n_j} (x_{i,j} - \overline X_j)^2$

$Q_2 = \sum \limits_{j=1}^k n_j (\overline X_j - \overline X)^2$

Нам в таблице дана внутригрупповая выборочная дисперсия

$\sigma^2_j = \frac{1}{n_j} \sum \limits_{i=1}^{n_j} (x_{ij} - \overline X_j)^2$, где $j$ это кол-во выборок, $n_j$ кол-во наблюдений в $j$-ой выборке.

Нам нужно вычислить $Q_1$, если мы домножим $\sigma^2_j$ на $n_j$, то получим

$n_j \cdot \sigma^2_j = n_j \frac{1}{n_j} \sum \limits_{i=1}^{n_j} (x_{ij} - \overline X_j)^2 = \sum \limits_{i=1}^{n_j} (x_{ij} - \overline X_j)^2$

тогда, если мы проссумируем откорректированную дисперсию по всем выборкам, получим статистику $Q_1$

$Q_1 = \sum \limits_{j=1}^k n_j \cdot \sigma^2_j = \sum \limits_{j=1}^k n_j \frac{1}{n_j} \sum \limits_{i=1}^{n_j} (x_{ij} - \overline X_j)^2 = \sum \limits_{j=1}^k \sum \limits_{i=1}^{n_j} (x_{ij} - \overline X_j)^2$

У нас все выборки равны между собой и равны 200.

```
nj = 200
Q1 = np.sum(nj * df['внутригрупповая выборочная дисперсия'].values)
Q1
```

Вычислим $Q_2 = \sum \limits_{j=1}^k n_j (\overline X_j - \overline X)^2$

```
Xmean_intergroup = df['внутригрупповое среднее'].values
num_group = len(Xmean_intergroup)
Xmean = np.sum(Xmean_intergroup) / num_group

Q2 = np.sum( nj * ((Xmean_intergroup - Xmean)**2) )
Q2, Xmean
```

Найдём статистику $F$

$F = \frac{Q_2 / (k-1)}{Q_1 / (N-k)}$, где

$k$ кол-во выборок, $N$ общее кол-во наблюдений всех выборок.

```
k = 5
N = nj * 5
numerator = Q2 / (k-1)
denominator = Q1 / (N-k)
Fv = numerator / denominator
Fv
```

Критическая область для $\alpha$:

$V_k = (u_{1 - \alpha, k - 1, N - k}; +\infty)$, где

$u_{1 - \alpha, k - 1, N - k}$ - квантиль Фишера.

$F_в \in V_k \Rightarrow H_0$ отвергается

```
alpha = 0.05
st.f.ppf(1 - alpha, k-1, N-k)
```

**Ответ:**

$F_в = 63.86$

$V_k = (2.38, + \infty)$

$F_в \in V_k \Rightarrow H_0$ отклоняется. Значит максимальная объемная скорость середины выдоха одинакова во всех группах.
