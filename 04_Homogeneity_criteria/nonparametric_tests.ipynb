В данном модуле будут рассмотрены критерии для проверки гипотезы об отсутсвии сдвига. 

Критерия называется непараметрическим, если закон распределения статистики критерия, независит от закона распределения исходной генеральной совокупности.

Рассмотрим две генеральные совокупности $\xi$ и $\eta$. Будем предполагать, что две сл. вел. $\xi$ и $\eta$ независимы.

Пусть ген. совокупность $\xi$ имеет ф-ю распределения $F(x)$ и из данной ген. совокупности получена выборка $X = (x_1, ..., x_n)$

$\xi, F(x), X = (x_1, ..., x_n)$

$\eta, G(x), Y = (y_1, ..., y_n)$

Требуется проверить гипотезу о том, что ф-и распределения совпадают.

$H_0: F(x) = G(x)$ для всех $x \in \mathbb {R}$

$H_1: F(x) \geq G(x)$ для всех $x \in \mathbb {R}$

$H'_1: F(x) \leq G(x)$ для всех $x \in \mathbb {R}$

$H''_2: F(x) \neq G(x)$ для всех $x \in \mathbb {R}$

Будем предполагать, что ф-и распр. $F(x)$ и $G(x)$ являются непрерывными, откуда следует, что эл-ты из выборок $X, Y$ не должны повторятся.

#### Непараметрические критерии для одномерных выборок

##### Критерий однородности Вилкоксона

Будем предполагать, что объем выборки $m$ из ген. совокупности $\eta$ не превосходит объема выборки $n$ из ген совокупности $\xi$.

Пусть $m \leq n$

Составим общий вариационный ряд из эл-ов выборки $X, Y$.

$ Z_{[n+m]} = (X_{[n]}, Y_{[m]}) $

$z_{(1)} < z_{(2)} < ... < z_{(m+n)}$

Назовём номер позиции эл-та выборки в вариационном ряду рангом. Найдём ранги элементов выборки $Y$. Через $s_1$ обозначим ранг эл-та $y_1$

$rank(y_1) = s_1$

Рассмотрим вычисления ранга на простом примере

$X_{[4]} = (8, 4, 6, 1)$

$Y_{[3]} = (0, 2, 3)$

Составим общий вариационный ряд, а под ним укажем ранг.

$0, 1, 2, 3, 4, 6, 8$

$1, 2, 3, 4, 5, 6, 7$

$rank(y_1) = 1$, $rank(y_2) = 3$, $rank(y_3) = 4$

$s_1 = 1$, $s_2 = 3$, $s_3 = 4$

Если сл. величины имеют одинаковые ф-ии распределения, в этом случае эл-ты выборок $X, Y$ будут равномерно перемешаны в вариационном ряду для $z$. 

Но если  $F(x) \geq G(x)$, тогда эл-ты выборки $X$ будут принимать значения меньшие, чем эл-ты выборки $Y$ и следовательно эл-ты выборки $Y$ будут сосредоточены в правой части вариационного ряда и ранги их будут больше.

Если $F(x) \leq G(x)$ эл-ты выборки $Y$ будут принимать меньшие значения, чем эл-ты выборки $X$ будут сосредоточены в левой части вариационного ряда для $z$ и ранги их будут в среднем меньше, чем ранги иксов.

###### Статистический критерий

В качестве статистики критерия, можно выбрать сумму рангов эл-ов выборки из генеральной совокупности $\eta$ (из выборки $Y$).

$W = \sum \limits_{i=1}^m s_i$

данная статистика принимает значения в интервале 

$[m (m+1)/2, mn + (m+1)m/2]$

###### Критическая область

Обратим внимание, что интервалы включают в себя крайние точки.

$H_1: F(x) \leq G(x)$

$V_k: [c_1, mn + (m+1)m/2]$, где

$c_1$ - некоторая критическая точка

$H_1: F(x) \geq G(x)$

$V_k: [m (m+1)/2, c_2]$

$H_1: F(x) \neq G(x)$

$V_k: [m (m+1)/2, c_3] \cup [c_4, mn + (m+1)m/2]$

Критические значения $c_1, c_2, c_3, c_4$ находят по таблицам критических значений статистики Вилкоксона или с помощью стат. пакетов, из условия

$P(W \in V_k|H_0) = \alpha$

если $W \in V_k \Rightarrow H_0$ отклоняется!

Полезные статьи: [студопедия](https://studopedia.ru/9_222473_kriteriy-vilkoksona.html), [блог](https://baguzin.ru/wp/neparametricheskie-kriterii-rangovy/)

Статистика критерия $W \in [m (m+1)/2, mn + (m+1)m/2]$

$DW=n∗m∗(m+n+1)/12$

```
alpha = 0.05 # уровень значимости
n, m = len(X), len(Y) # объемы выборок
mt = n*(n+m+1)/2 # мат. ожидание W при H0
se = np.sqrt(m*n*(m+n+1)/12)# дисперсия W при H0
zkp1 = st.norm.ppf(alpha/2)# z - критическое значение (левое)
zkp2 = st.norm.ppf(1-(alpha/2))# z - критичесокое значений (правое)
# zkp2 = (n + m + 1) * n - wkp1 # альтернативно, можно вычислить по этой формуле https://studopedia.ru/9_222473_kriteriy-vilkoksona.html
Wdown = mt + zkp1*se
Wup = mt + zkp2*se
print(Wdown, Wup)
```

###### Апроксимация

В случае, когда объемы выборок велики, мы можем воспользоваться следующей апроксмацией, мы можем центрировать и нормировать статистику $W$

$ \tilde W = \frac{W - \frac{m(m+n+1)}{2}}{\sqrt{ \frac{mn(m+n+1)}{12} }} \xrightarrow[m,n \rightarrow \infty]{H_0 \; d} N(0, 1)$

Для больших объемов выборок, мы можем построить приближенный критерий

$H_0: F(x) = G(x)$

$H_1: F(x) \neq G(x)$

нам достаточно найти квантиль стандартного нормального распределения

$u_{1 - \frac{\alpha}{2}}$

и посмтроить критическую область

$V_k = (|\tilde w_в| > u_{1 - \frac{\alpha}{2}})$

$\tilde w_в \in V_k \Rightarrow H_0$ отклоняем

Если у нас большие выборки, то могут появится одинаковые значения, например, из-за округления. В данном случае необходимо произвести поправку для $\tilde W$ в знаменателе. 

В знаменатель ставим такое выражение

$\sqrt{ \frac{mn(m+n+1)}{12} \bigg (1 - \frac{\sum \limits_{i=1}^k t_i (t_i^2 - 1)}{(n+m)(n+m-1)(n+m+1)} \bigg ) }$

$\tilde W =  \frac{W - \frac{m(m+n+1)}{2}}{\sqrt{ \frac{mn(m+n+1)}{12} \bigg (1 - \frac{\sum \limits_{i=1}^k t_i (t_i^2 - 1)}{(n+m)(n+m-1)(n+m+1)} \bigg ) }}$

$k$ число связок, т.е. наборов повторяющихся значений

$t_i$ - размер $i$-й связки, учитываются только те связки, в которые входят эл-ты из разных выборок. если в связку входят эл-ты из одной и той же выборке, в этом случае они не влияющт на статистику $\tilde W$

#### Непараметрические критерии для парных выборок

Сл. вел. зависимы, т.к. измерения производились на одном и том же объекте.

На практике мы прибегаем к данному критерию, в случае, когда два измерения производились на одних и тех же объектах. Например, измерили температуру у одних и тех же людей до и после приема лекарства.

Получена двумерная выборка $(\eta, \xi)$, $(X, Y)$.

Требуется проверить гипотезу о равенстве ф-й распределений, т.е. мы проверяем, есть ли изменения или нет.

$H_0: F(x) = G(x)$ для всех $x \in \mathbb {R}$

$H_1: F(x) \leq G(x)$ для всех $x \in \mathbb {R}$

$H'_1: F(x) \geq G(x)$ для всех $x \in \mathbb {R}$

$H''_1: F(x) \neq G(x)$ для всех $x \in \mathbb {R}$


т.к. наши сл. величины связаны, рассмотрим новую сл. вел. $\zeta$

$\zeta = \eta - \xi$

$z_i = x_i - y_i$

В случае когда справедлива $H_0$ гипотеза, события $(z_i < 0)$ и $(z_i > 0)$ равновероятны, если у нас выборки взяты их непрерывной ген. совокупности, в этом случае данные вероятности будут равны $\frac{1}{2}$.

Поэтому давайте вместо гипотезы $H_0: F(x) = G(x)$ будем рассматривать гипотезу 

$H_0: P(z_i < 0) = P(z_i > 0) = \frac{1}{2}$

$H_1: P(z_i < 0) > P(z_i > 0)$

$H'_1: P(z_i < 0) < P(z_i > 0)$

$H''_1: P(z_i < 0) \neq P(z_i > 0)$

Важно!!! Из равенства $F(x) = G(x) \Rightarrow P(z_i < 0) = P(z_i > 0)$ обратное не верно!

Гипотеза о равенстве вероятности, скорее гипотеза о равенстве медиан двух генеральных совокупностей.

##### Критерий знаков (биноминальный критерий)



###### Критическая область

###### Статистический критерий 

$L = \sum I(z_i < 0) \sim Bin(n, \frac{1}{2})$ в случае, когда справедлива $H_0$

Статистика принимает значения из интервала

$L \in [0, n]$

$H_1: F(x) \geq G(x)$

$V_k: (c_1, n]$

$H_1: F(x) \leq G(x)$

$V_k: [0, c_2)$

$H_1: F(x) \neq G(x)$

$V_k: [0, c_3) \cup (c_4, n]$

Критические значения $c_1, c_2, c_3, c_4$ находят по таблицам вероятностей биноминального распределения, в предположении, что справедлива $H_0$ и $P(z_i < 0) = \frac{1}{2}$, для этого нам нужно будет ещё знать объем выборки $n$ или с помощью стат. пакетов, из условия

$P(L \in V_k|H_0) = \alpha$



###### Scipy

Найти $c_i$ можно в пакете $scipy.stats.binom.ppf(alpha, n, p)$, где

$alpha$ - уровень значимости

$n$ - кол-во выборки

$p$ - параметр вер-ти биноминального распределения (p=0.5)

если $L \in V_k \Rightarrow H_0$ отклоняется!

##### Критерий знаковых ранговых сумм Вилкоксона

Статистика основана на рангах модулей $z_i$. Рассмотрим выборку $z_1, ..., z_n$, где $z_i = x_i - y_i$.

На основе её построим последовательность модулей

$|z_1|, ..., |z_n|$

и упорядочим их

$|z'_1| < ... < |z'_n|$

Найдём ранги модулей $|z_i|$ в этой последовательности

$s_1 = rank(|z_1|), ..., s_n=rank(|z_n|)$

т.е. берем модуль $|z_1|$ и смотрим, в какой позиции упорядоченного ряда он находится. $s_i$ ранг модуля $|z_i|$

###### Задача №2

Если смотреть на луну которая находится на горизонте, она кажется по размеру боьлше, чем луна находящаяся в зените.

В 1940 г. Халвой и Боринг предположили, что это связано с тем, что в одном случае человек смотрит на луну прямо, а в другом случае ему приходится поднимать глаза или голову. Они решили проверить эту гипотезу и провели следующий эксперимент.

В эксперименте участвовало 10 испытуемых. У них была аппаратура, которая позволяла показывать две искуственных луны, одну на уровне горизонта, другую в зените. А так же следить за тем, как смотрят на луну испытуемый, прямо или ему приходится поднимать взгляд. Каждый испытуемый смотрел на две луны. На луну находящуюся на горизонте, он смотрел прямо, а на луну в зените один раз смотрел прямо, другой раз поднимая взгляд. Испытуемые должны были сказать ощущаемые размеры луны. В итоге сравнивались значения отношений разера в зените и горизонте. Было получено 10 парных наблюдений, каждая пара соотвествует определенному испытуемому. 

на уровне X 1.65, 1.0, 2.04, 1.25, 1.05, 1.02, 1.67, 1.86, 1.56, 1.73

выше Y 1.73, 1.06, 2.03, 1.4, 0.95, 1.13, 1.41, 1.73, 1.63, 1.56





$H_0: F(x) = G(x)$ для всех $x \in \mathbb R$

$H_1: F(x) \leq G(x)$ для всех $x \in \mathbb R$

Учёные предположили, что луна больше на горизонте, потому что люди смотрят на неё прямо, если следовать этой гипотезе, тогда значения должны быть

$x_i > y_i$

Поэтому мы выбрали такую альтернативную гипотезу $H_1$.

Проверим гипотезу используя критерий знаковых ранговых сумм Вилкоксона.

$U = \sum I(z_i < 0) \cdot s_i$, где

$s_i$ ранг, модуля статистики $Z = X - Y$



```
X = np.array([1.65, 1.0, 2.04, 1.25, 1.05, 1.02, 1.67, 1.86, 1.56, 1.73])
Y = np.array([1.73, 1.06, 2.03, 1.4, 0.95, 1.13, 1.41, 1.73, 1.63, 1.56])
# таким способом можно найти статистику U, но только у нас альтернатива левосторонняя, а тут стоит двухсторонняя. 
# ответ верный, но возможно это совпадение
U, p = st.wilcoxon(X, Y, alternative='two-sided')
U
```

Построим оценку вручную.

Найдём $Z = X - Y$

Нам необходимо найти модули $z_i$ и упорядочить их по возрастанию.

```
Z = X - Y
z_rank = np.sort(abs(Z))
# отберем отрицательные значения, ранги которых нужно найти, возьмем от них модуль, т.к. в z_rank у нас модули
search = abs(Z[Z < 0])
# найдём индексы отрицательных значений, т.к. индексация идёт с 0, а нам нужны ранги с индексацией с 1, мы добавим единицу
U = (np.searchsorted(z_rank, search) + 1).sum()
U
```

Найдём критическое значение $W_{\alpha, n}$ с уровнем значимости $\alpha$ и кол-ом выборки $n$ из [таблицы](https://hostingkartinok.com/show-image.php?id=e8485e09380e2c3b479bbbaca8720f5a).

$\alpha = 0.05$

$n = 10$

$W_{\alpha, n} = 11$

Тогда критическая область для $H_1: F(x) \leq G(x)$ будет

$V_k: [0, 11)$

$U \notin V_k \Rightarrow H_0$ не отвергается.

**Ответ:**

$H_0$ не отвергается.
